```python
# Cell 1: Setup & Download
import pandas as pd
import requests, zipfile, io, joblib
from pathlib import Path
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

Path("data/raw").mkdir(parents=True, exist_ok=True)
Path("models").mkdir(exist_ok=True)

url = "https://archive.ics.uci.edu/static/public/162/forest+fires.zip"
r = requests.get(url)
with zipfile.ZipFile(io.BytesIO(r.content)) as zf:
    zf.extractall("data/raw")
```

```python
# Cell 2: Training & Evaluasi
df = pd.read_csv("data/raw/forestfires.csv")
features = ['temp', 'RH', 'wind', 'rain', 'DMC', 'DC', 'ISI']
target = 'area'

X = df[features]
y = df[target]

model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X, y)

# Evaluasi Model
y_pred = model.predict(X)
print(f"R-Squared Score: {r2_score(y, y_pred):.2f}") # Dosen S2 akan melihat ini!

# Simpan
joblib.dump(model, "../models/fire_model.pkl")
joblib.dump(features, "../models/feature_names.pkl")
```
